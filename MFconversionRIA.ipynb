```python
# import required data packages / fuzzywuzzy also imported below
import pandas as pd
import numpy as np
# define a dataframe with location and type of file, skipping first 3 rows of fund director, and selecting desired tab
# clean data by remove NAN and NA on fund name
df = pd.read_excel('\\Users\\Brett\\OneDrive\\Desktop\\MF Conversion\\IFN_directory_RIA.xlsx', skiprows=2, sheet_name ="IFN_directory_RIA").dropna(subset=['Fund Name'])
# display 4 rows to test
df.head()
```

```python
# import tickers from advisor file called advisorData
advisorData = pd.read_excel('\\Users\\Brett\\OneDrive\\Desktop\\MF Conversion\\riaData\\advisorData.xlsx')
advisorData
```

```python
# find matches to gather fund data
dffunds = df.loc[df['Symbol'].isin(advisorData['VHIAX'
])]
dffunds
```

```python
#find partial matches
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
matches_list = []

for row in df.index:
    co_name = df.get_value(row, "Fund Name")
    expRatio = df.get_value(row, "Expense Ratio")
    shClass = df.get_value(row, "Share Class")
    ticker = df.get_value(row, "Symbol")
    ffname = df.get_value(row, "Fund Family Name")
    for columns in dffunds.index:
        aco_name = dffunds.get_value(columns, "Fund Name")
        ffMatch = dffunds.get_value(columns, "Fund Family Name")
        matched_token=fuzz.partial_ratio(co_name,aco_name)
        if matched_token >67:
            matches_list.append([ffname,aco_name, ffMatch, co_name, matched_token, expRatio, shClass,ticker])
            
```

```python
print(*matches_list)
```

```python
df2 = pd.DataFrame(matches_list, columns= ['Fund Family Name','Fund Name', 'Fund Family Name1', 'Fund Name1', 'Match percent', 'Expense Ratio', 'Share Class','Ticker'])
```

```python
#only pulling data where the fund family names are the same for each match
df2 = df2.loc[df2['Fund Family Name'].isin(df2['Fund Family Name1'
])]
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter("MatchesList.xlsx", engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.

df2.to_excel(writer, sheet_name='Sheet1')

# Get the xlsxwriter workbook and worksheet objects.
workbook  = writer.book
worksheet = writer.sheets['Sheet1']

# Add some cell formats.
format1 = workbook.add_format({'num_format': '#,##0.00'})
format2 = workbook.add_format({'num_format': '0%'})

# Note: It isn't possible to format any cells that already have a format such
# as the index or headers or any cells that contain dates or datetimes.

# Set the column width and format.
worksheet.set_column('B:B', 18, format1)

# Set the format but not the column width.
worksheet.set_column('C:C', None, format2)

worksheet.autofilter('B1:E1000')
# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
